```python
import pyvista as pv

# Define the path for the .vtr file
vtr_file_path = 'ddprocessoutput\VTRoutput_1.vtr'

# Read the .vtr file using PyVista
grid = pv.read(vtr_file_path)

# Extract the grid points and a sample scalar/vector field for visualization
scalar_field_name = grid.array_names[0] if grid.array_names else None
scalar_field = grid[scalar_field_name] if scalar_field_name else None

# 3D visualization
plotter = pv.Plotter()
plotter.add_mesh(grid, scalars=scalar_field, scalar_bar_args={"title": scalar_field_name}, opacity=0.8)
plotter.show()
```
